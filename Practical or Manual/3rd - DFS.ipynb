```python
# from backside || Working

# Sample Graph 1
# graph = {
#   '2' : [],
#   '3' : ['2', '4'],
#   '4' : ['8'],
#   '5' : ['3','7'],
#   '7' : ['8'],
#   '8' : []
# }

# Sample Graph 2
graph = {
  '1' : [],
  '2' : [],
  '4' : ['5','6'],
  '5' : ['7'],
  '6' : ['1','2'],
  '7' : []
}

# Sample view as node of graph
#        4
#    5       6
# 7       1     2


visited = []   # List for visited nodes.
stack = []     # Initialize a stack

def dfs(visited, stack, node): #function for DFS
  visited.append(node)
  stack.append(node)

  while stack:          # Creating loop to visit each node
    m = stack.pop()
    print (m, end = " ")

    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        stack.append(neighbour)

# Function call
print("\nFollowing is the Depth-First Search")
dfs(visited, stack, '4')
```
