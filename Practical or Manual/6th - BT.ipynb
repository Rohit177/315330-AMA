**Example:**
<br>
> **Disease test:** 1 in 1000 people have disease.

> 99% Accurate (true positive rate = 0.99).

> False positive rate is 5%.

````python
def bayes_theorem(pa, pba, pb):
    """
    Bayes' Theorem:
    P(A|B) = (P(B|A) * P(A)) / P(B)
    """
    pab = (pba * pa) / pb
    return pab


# Given:
pa = 0.01         # P(A)
pba = 0.95  # P(B|A)
pp = (pba * pa) + (0.05 * (1 - pa))  # Total positive

# Calculate
result = bayes_theorem(pa, pba, pp)

print(f"Positive test of disease: {result:.3f}")
````

````python
# prompt: create bayes theoram

# Example:
# <br>
# > **Disease test:** 1 in 100 people have disease.
#
# > 95% Accurate (true positive rate = 0.95).
#
# > False positive rate is 5%.
def bayes_theorem(pa, pba, pb):
    """
    Bayes' Theorem:
    P(A|B) = (P(B|A) * P(A)) / P(B)
    """
    pab = (pba * pa) / pb
    return pab


# Given:
pa = 0.01         # P(A) - Probability of having the disease
pba = 0.95  # P(B|A) - Probability of a positive test given the disease (True Positive Rate)
# Calculate P(B) - Probability of a positive test
# P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
# P(not A) = 1 - P(A)
# P(B|not A) = False Positive Rate = 0.05
pb = (pba * pa) + (0.05 * (1 - pa))

# Calculate P(A|B) - Probability of having the disease given a positive test
result = bayes_theorem(pa, pba, pb)

print(f"Probability of having the disease given a positive test: {result:.3f}")
```
````
