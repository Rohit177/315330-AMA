```python
# prompt: create a code where 1 is start 7 is goal and best bath with bfs

def bfs_shortest_path(graph, start, goal):
    # Keep track of visited nodes
    visited = []
    # Keep track of all the paths to be checked
    queue = [[start]]

    # Return path if start is goal
    if start == goal:
        return [start]

    # Loop while there are paths in the queue
    while queue:
        # Pop the first path from the queue
        path = queue.pop(0)
        # Get the last node from the path
        node = path[-1]
        # If the node is not visited
        if node not in visited:
            # Add the node to visited list
            visited.append(node)
            # Get the neighbors of the current node
            neighbors = graph.get(node, [])

            # Go through all neighbor nodes
            for neighbor in neighbors:
                # Create a new path by adding the neighbor
                new_path = list(path)
                new_path.append(neighbor)
                # If neighbor is the goal, return the new path
                if neighbor == goal:
                    return new_path
                # Otherwise, add the new path to the queue
                queue.append(new_path)

    # Return None if no path is found
    return None

# Define the graph based on the provided context
graph = {
  '1' : ['2','5'],
  '2' : ['3','7'],
  '3' : ['4'],
  '4' : ['4'],
  '5' : ['6'],
  '6' : [],
  '7' : ['6']
}

# Define start and goal
start_node = '1'
goal_node = '6'

# Find the shortest path using BFS
shortest_path = bfs_shortest_path(graph, start_node, goal_node)

# Print the result
if shortest_path:
    print(f"Shortest path from {start_node} to {goal_node}: {' -> '.join(shortest_path)}")
else:
    print(f"No path found from {start_node} to {goal_node}")

# Test with another path
start_node_2 = '1'
goal_node_2 = '4'
shortest_path_2 = bfs_shortest_path(graph, start_node_2, goal_node_2)
if shortest_path_2:
    print(f"Shortest path from {start_node_2} to {goal_node_2}: {' -> '.join(shortest_path_2)}")
else:
    print(f"No path found from {start_node_2} to {goal_node_2}")
```

```python
graph = {
  '1' : ['2','5'],
  '2' : ['3','7'],
  '3' : ['4'],
  '4' : ['7'],
  '5' : ['6'],
  '6' : [],
  '7' : ['6']
}


def dfs(graph, start, goal):
    stack = [(start, [start])]  # Stack holds (current_node, path_so_far)

    while stack:
        (node, path) = stack.pop()
        if node == goal:
            return path  # Goal found, return path

        for neighbor in reversed(graph[node]):  # Reverse to mimic traditional DFS order
            if neighbor not in path:  # Avoid revisiting
                stack.append((neighbor, path + [neighbor]))

    return None  # No path found

start_node = '1'
goal_node = '6'

bfs_shortest_path(graph, start_node, goal_node)
```
